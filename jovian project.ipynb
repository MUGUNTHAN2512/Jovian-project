## Exploratory Analysis and Visualization

In this section I am going to analyze and visualize some datas




Let's begin by importing`matplotlib.pyplot` and `seaborn`.

````python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

#1 Which country players has enrolled maximum in the auction?

First we were finding how many players are in the auction in the given data.
next using value_counts() finding which country players have enrolled in auction.

````python
top_10_countries = auction_df.Country.value_counts().head(10)
top_10_countries
````

````python
plt.figure(figsize=(16,9))
plt.title('Top 10 Countries')
sns.barplot(x = top_10_countries, y = top_10_countries.index)
plt.show;
````

#2 - PIE CHART VISUALIZATION FOR SPECIALISM OF PLAYERS PARTICIPATING IN AUCTION

I am going to do visualization for specialism of participation in auction like "All Rounder","Wicket Keeper","Bowler",etc


````python
plt.title('Specialism of players')
plt.pie(auction_df['Specialism'].value_counts(),labels=auction_df['Specialism'].unique(),autopct='%2.1f%%',startangle=0);
````

#3 - PIE CHART FOR SOLD AND UNSOLD PLAYERS IN THIS AUCTION

I am going to do for the player is sold or unsold in this auction.

````python
plt.title('sold Vs unsold')
plt.pie(auction_df['Bid'].value_counts(),labels=auction_df['Bid'].unique(),autopct='%2.1f%%',startangle=0);
````

#4 HOW MANY PLAYERS ARE THERE WITH DIFFERENT TYPES OF BOWLING STYLE 

here there is a clear visualization for the Bowling style of the players using countplot

````python
plt.title('Bowling style')
sns.countplot(y = 'Bowling Style', data = auction_df);
````

#5 HOW MANY NEW PLAYERS ARE SELECTED BY NEW IPL TEAM

here i have found how many players are sold in the IPL Auction 2022

````python
ipl_team = auction_df.Team.value_counts().drop("UnSold")
ipl_team
````

````python
plt.figure(figsize=(16,9))
plt.title('Top 10 Countries')
sns.barplot(x = ipl_team, y = ipl_team.index)
plt.show;
````

## Asking and Answering Questions

In this section, We have to ask and answer the questions. Each of the question will be solved by using the libraries like
pandas, matplotlib, seaborn, numpy. I'm going to ask the interesting questions related to the data as well as going

Q 1: WHICH PLAYER SOLD FOR MAX PRICE

lets find how many players are sold for maximum price in this auction

````python
players = auction_df.sort_values('Price Paid', ascending = False)
players
````

````python
top_10_paid_players = players[['Players', 'Price Paid']].head(10)
top_10_paid_players
````

````python
plt.figure(figsize=(16,9))
plt.title('Top 10 players')
plt.xticks(rotation = 90)
sns.barplot(x = 'Players', y = 'Price Paid', data = top_10_paid_players)
plt.show;
````

 Q2: HOW MANY INDIAN PLAYERS SOLD IN THE AUCTION
 
 calculation of sold indian players

````python
ind_play = auction_df.loc[(auction_df['Country'] == 'India') & (auction_df['Bid'] == 'Sold')]
ind_play
````

````python
indian_players = ind_play.Players.count()
indian_players
````

 Q3: HOW MANY CAPPED PLAYERS WERE SOLD IN THE AUCTION
 
 let us find how many capped players are sold in this auction

````python
cap_play = auction_df.loc[(auction_df['C/U/A'] == 'Capped') & (auction_df['Bid'] == 'Sold')]
cap_play
````

````python
capped_players = cap_play.Players.count()
capped_players
````

Q4: LIST OUT THE PLAYERS OF RAJASTHAN ROYALS

let us find how many players are selected for Rajastan Royals in thia auction

````python
rr_players = auction_df[auction_df['Team'] == 'RR']
rr_players
````

````python
total_rr_players = rr_players.Players.count()
total_rr_players
````

 Q5: WHICH ALL-ROUNDER PLAYERS HAS HIGHEST TEST CAPS

````python
all_rounder = auction_df[auction_df['Specialism'] == 'ALL-ROUNDER']
all_rounder
````

````python
top_10 = all_rounder[['Players','Test caps']].head(10)
top_10
````
